## 2.1 Data Cleaning <a id="part2.1"></a>
[Back to top](#top)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cardio_good_fitness.csv')
```

```python
df_bkup = df.copy()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
print("Number of blanks")
print("Age:", 183 - df['Age'].count())
print("Usage:", 183 - df['Usage'].count())
print()
print("Number of na")
print("Product:", (df['Product'] == 'na').sum())
print("Branch:", (df['Branch'] == 'na').sum())
print("Age:", (df['Age'] == 'na').sum())
print("Gender:", (df['Gender'] == 'na').sum())
print("Education:", (df['Education'] == 'na').sum())
print("MaritalStatus:", (df['MaritalStatus'] == 'na').sum())
print("Usage:", (df['Usage'] == 'na').sum())
print("Fitness:", (df['Fitness'] == 'na').sum())
print("Income:", (df['Income'] == 'na').sum())
print("Miles:", (df['Miles'] == 'na').sum())
print()
print("Marital Status")
print("S:", (df['MaritalStatus'] == 'S').sum())
print("P:", (df['MaritalStatus'] == 'P').sum())
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace({'P': 'Partnered', 'S': 'Single'})
```

```python
df['Age'] = df['Age'].replace('na', np.nan)
df['Age'] = pd.to_numeric(df['Age'])
mean_age = int(df['Age'].mean())
df['Age'] = df['Age'].fillna(mean_age)
df
```

```python
df['Usage'] = df['Usage'].replace('na', np.nan)
df['Usage'] = pd.to_numeric(df['Usage'])
mean_usage = int(df['Usage'].mean())
df['Usage'] = df['Usage'].fillna(mean_usage)
df
```

```python
df = df.replace("na", np.nan)
df = df.dropna()
df = df.reset_index(drop=True)
df
```

```python
df["Age"] = df["Age"].astype(int)
df['Education'] = df['Education'].astype(int)
df["Usage"] = df["Usage"].astype(int)
df["Fitness"] = df["Fitness"].astype(int)
df["Income"] = df["Income"].astype(int)
df["Miles"] = df["Miles"].astype(int)
df.info()
```

```python
print("Duplicates: ", df.duplicated().sum())
```

## 3.1 What is the proportion of customers who purchased each treadmill model? <a id="part3.1"></a>
[Back to top](#top)

```python
product_counts = df['Product'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(product_counts.index, product_counts.values, color=['b','r','g'])

ax.set_title('Proportion of Customers by Treadmill Model Purchased', fontsize=14)
ax.set_ylabel('Number of Purchases', fontsize=12)
ax.set_xlabel('Treadmill Model', fontsize=12)

for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 0), textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()

plt.show()
```

The bar chart shows the number of purchases for each treadmill model TM195, TM498, and TM798. The height of each bar indicates how many customers selected that particular model. It is evident that TM195 has the highest number of purchases at 80, making it the most popular choice among the three. TM498 follows with 60 purchases, suggesting that it still attracts a substantial share of customers, but not as many as TM195. TM798, at 40 purchases, has the lowest uptake, implying it appeals to a smaller segment of the market.
From these figures, one can see that TM195 constitutes approximately 44% of the total 180 purchases (80 out of 180), TM498 makes up around 33% (60 out of 180), and TM798 accounts for roughly 22% (40 out of 180). In my view, this difference could be driven by factors such as price, features, or brand perception, with TM195 appearing to resonate most strongly with buyers. 

This chart addresses the exploratory question “What is the proportion of customers who purchased each treadmill model?” by highlighting that TM195 dominates overall sales, followed by TM498, while TM798 remains less commonly chosen.

## 3.4 Do men and women show different preferences when purchasing treadmill models? <a id="part3.4"></a>
[Back to top](#top)

```python
x_labels = df["Product"].unique()
x = np.arange(len(x_labels))

men_counts = df[df["Gender"] == "Male"]["Product"].value_counts().reindex(x_labels, fill_value=0)
women_counts = df[df["Gender"] == "Female"]["Product"].value_counts().reindex(x_labels, fill_value=0)

width = 0.35  

fig, ax = plt.subplots(figsize=(8, 5))

bar1 = ax.bar(x - width/2, men_counts, width, label='Male', color='c')
bar2 = ax.bar(x + width/2, women_counts, width, label='Female', color='m')

ax.set_title("Treadmill Purchases by Gender", fontsize=20)
ax.set_xlabel("Treadmill Model", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend()

def autolabel(bars):
    for each in bars:
        height = each.get_height()
        ax.annotate('{}'.format(height),
                    xy=(each.get_x() + each.get_width() / 2, height),
                    xytext=(0, 2),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)

fig.tight_layout()

plt.show()
```

The grouped bar chart shows the number of treadmills purchased by males and females across three different treadmill models TM195, TM498, and TM798. The cyan bars represent male buyers, whereas the magenta bars represent female buyers. It is evident that for TM195, both men and women purchased this model in equal numbers, with 40 purchases each, indicating a balanced preference. In contrast, TM498 shows a slight male preference, with more men purchasing this model compared to women, although the difference is minimal. The biggest difference is observed in TM798, where a lot more men purchased this model compared to women. While male purchases for TM798 remain high at 33 units, female purchases are considerably lower, at 7 units. This suggests that TM798 is more popular among men. It can be concluded that while men and women show similar preferences for TM195 and TM498, there is a big contrast in TM798 purchases, where men dominate. This could be due to factors such as pricing, advanced features, or marketing appeal that make TM798 more attractive to male buyers. 

This analysis answers the question, “Do men and women show different preferences when purchasing treadmill models?” by revealing that while some models are equally popular among both genders, TM798 is strongly favoured by men over women.

## 3.7 Which treadmill model would a customer get if their usage will be high? <a id="part3.7"></a>
[Back to top](#top)

```python
plt.figure(figsize=(8, 5))

colour_mapping = {"TM195": "b", "TM498": "r", "TM798": "g"}

usage_trend = df.groupby(["Usage", "Product"], observed=False).size().unstack()

for product in usage_trend.columns:
    plt.plot(
        usage_trend.index, 
        usage_trend[product], 
        marker='o', 
        label=product, 
        color=colour_mapping.get(product)
    )

plt.title("Trend of Weekly Treadmill Usage Across Different Models", fontsize=14)
plt.xlabel("Weekly Usage (Times per Week)", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.legend(title="Treadmill Model")

plt.tight_layout()

plt.show()
```

The line chart shows the relationship between how often customers use their treadmill per week and the number of customers who follow that usage pattern. The three treadmill models, TM195, blue, TM498, red, and TM798, green, are represented by different lines. From the graph, it is clear that TM195 and TM498 are most commonly used by customers who use their treadmill 2 to 3 times per week. TM195 peaks at 3 times per week with the highest number of users, around 37 customers, while TM498 has a similar trend, reaching its highest usage at 3 times per week, about 30 customers,. However, as weekly usage increases to 4 or more times per week, the number of users for TM195 and TM498 drops sharply. TM798, on the other hand, shows a different trend. While it has fewer users overall, its popularity increases as weekly usage rises. The number of TM798 users peaks around 4 to 5 times per week, and it continues to have steady usage even at 6 to 7 times per week, whereas the other models decline. This suggests that customers who plan to use their treadmill frequently are more likely to choose TM798.

This chart helps answer the question “Which treadmill model would a customer get if their usage will be high?” by showing that TM798 is the best choice for customers with high weekly usage. The other models, TM195 and TM498, are more popular among those with lower to moderate treadmill usage, typically around 2 to 3 times per week. TM798’s ability to maintain a steady number of users at 5 or more times per week suggests it may be more durable or better suited for frequent usage.

## 3.11 How is treadmill usage affected by both fitness levels and income? <a id="part3.11"></a>
[Back to top](#top)

```python
fitness_levels = np.sort(df["Fitness"].unique())
usage_by_fitness = df.groupby("Fitness")["Usage"].mean()
income_by_fitness = df.groupby("Fitness")["Income"].mean()

fig, ax_u = plt.subplots(1, 1, figsize=(8, 4))

ax_u.bar(fitness_levels, usage_by_fitness, width=0.4, label='Avg Weekly Usage', color='c')
ax_u.set_ylim(bottom=0, top=max(usage_by_fitness) + 2)
ax_u.set_title("Fitness Level on Treadmill Usage and Income", fontsize=16)
ax_u.set_xlabel("Fitness Level", fontsize=12)
ax_u.set_ylabel("Average Weekly Usage", fontsize=12, color='c')

ax_i = ax_u.twinx()

ax_i.plot(fitness_levels, income_by_fitness, 'g-o', label='Avg Income')
ax_i.set_ylim(bottom=min(income_by_fitness) * 0.9, top=max(income_by_fitness) * 1.1)
ax_i.set_ylabel("Average Income ($)", fontsize=12, color='g')

ax_u.grid(False)

fig.legend(loc=(0.1, 0.7))

fig.tight_layout()

plt.show()
```

The dual axis chart examines the relationship between fitness levels, treadmill usage, and income. The x-axis represents the fitness level, from 1 = least fit to 5 = very fit.
